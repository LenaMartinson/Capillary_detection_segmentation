## Make plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Graphic of comparison of image sizes

# Labels to show on graphic
names = [
    "320",
    "640",
    "960",
    "1280"
]

# Results (sharp precision)
results_train = [0.5338383940879636, 0.6183332760654712, 0.6484489628244566, 0.6700976841152929]
results_val = [0.40326613124743493, 0.4502607877866136, 0.4619244987463018, 0.45784532409796924]
results_test = [0.5167525086730248, 0.5411444777768473, 0.5567403979489225, 0.5768873985258105]

# Making a plot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 8))
plt.title("Sizes comparison for YOLOv8", fontsize=15)

plt.plot(names, results_train,  '-ok', color='red', label='train')
plt.plot(names, results_val, '-ok', color='blue', label='val')
plt.plot(names, results_test, '-ok', color='green', label='test')

plt.xlabel("imgsz")
plt.ylabel("Precision (sharp)")

plt.legend(fontsize=14)
plt.savefig("sizes_2.png")
```
