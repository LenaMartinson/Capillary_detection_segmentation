## How to calculate metrics

```python
# Upload YOLOv8 weights (last)
model = YOLO('/home/martinson/runs/segment/seg; train s; small_data; no_aug _tune mech3/weights/last.pt')

# Set params
params = {
    "conf":0.1,
    "iou":0.1
}

# Evaluate results on valation and test 
results_val = model.val(**params)
results_test = model.val(split='test', **params)

# Upload YOLOv8 weights (best)
model = YOLO('/home/martinson/runs/segment/seg; train s; small_data; no_aug _tune mech3/weights/best.pt')

# Evaluate results on valation and test 
results_val_2 = model.val(**params)
results_test_2 = model.val(split='test', **params)
```

## Calculate results for YOLOv5

You need to change YOLOv5 code a bit to make model return precision

```python
import yolov5.segment.val as val

folder_name = "/home/martinson/yolov5/runs/train-seg/seg; train s; no_aug/weights/"

# evaluate precision (SHARP) -- need to change YOLOv5 code for it!
_, p_train = val.run(
    data="...",
    weights=folder_name + "best",
    iou_thres=0.1,
    task='train'
)

_, p_val = val.run(
    data="...",
    weights=folder_name + "best",
    iou_thres=0.1,
    task='val'
)
_, p_test = val.run(
    data="...",
    weights=folder_name + "best",
    iou_thres=0.1,
    task='test'
)
```

```python
# 5
# seg; train s; default

print("Best")
print("Train (sharp precision):", p_train)
print("Val (sharp precision):", p_val)
print("Test (sharp precision):", p_test)
```
